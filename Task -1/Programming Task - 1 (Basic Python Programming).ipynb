## Problem-4: Python If-Else

**Task**
Given an integer `n`, perform the following conditional actions:

* If `n` is odd, print `Weird`
* If `n` is even and in the inclusive range of 2 to 5, print `Not Weird`
* If `n` is even and in the inclusive range of 6 to 20, print `Weird`
* If `n` is even and greater than 20, print `Not Weird` 
**Input Format**
A single line containing a positive integer, `n`. 

**Constraints**

* `n` is a positive integer. 

**Output Format**
Print `Weird` or `Not Weird` based on the conditions above. 
**Sample Input 0**

```
3
```

**Sample Output 0**

```
Weird
```

Explanation: `n = 3` is odd → print `Weird`. 

**Sample Input 1**

```
24
```

**Sample Output 1**

```
Not Weird
```

Explanation: `n = 24` is even and > 20 → print `Not Weird`. 

`````python
n = int(input())

if n % 2 == 1:
    print("Weird")
elif 2 <= n <= 5:
    print("Not Weird")
elif 6 <= n <= 20:
    print("Weird")
else:
    print("Not Weird")
`````

## Problem-5: Loops

**Task**
Given an integer `n`, print the squares of all non-negative integers less than `n`.
For every integer `i` in the range `0` to `n-1`, print:

```
i*i
```

Each result should appear on its own line.

**Input Format**
A single integer `n`.

**Constraints**

* `1 ≤ n ≤ 20`

**Output Format**
Print `i*i` for every `i` in `[0, n)`.

**Sample Input**

```
5
```

**Sample Output**

```
0
1
4
9
16
```



`````python
n = int(input())

for i in range(n):
    print(i**2)  #For Loop
`````

`````python
n = int(input())

i = 0
while i < n:
    print(i * i)
    i += 1      #While loop 
`````

## Problem-7: Write a Function

### Task

Given a year, determine whether it is a leap year. If it is a leap year, return the Boolean `True`, otherwise return `False`.

A leap year in the Gregorian calendar satisfies:

* The year can be evenly divided by 4 — **a leap year**, unless:
* The year can be evenly divided by 100 — **not a leap year**, unless: ([HackerRank][1])
* The year is also evenly divisible by 400 — **a leap year**.

Thus:

* Years like 2000 and 2400 are leap years.
* Years like 1800, 1900, 2100, 2200, 2300, and 2500 are *not* leap years. 

---

### Input Format

A single integer, `y`, denoting the year. 

### Constraints

`1900 ≤ y ≤ 10**5` 

---

### Output Format

Your function must return a Boolean value: `True` if `y` is a leap year, `False` otherwise. The code stub handles input/output. 

---

### Sample Input

```
1990
```

### Sample Output

```
False
```

### Explanation

1990 is not evenly divisible by 4, so it is not a leap year. 


`````python
def is_leap(year):
    if (year%400 == 0) or (year%4 == 0 and year%100 != 0):
        Leap = True
    else:
        Leap = False
    return Leap

year = int(input())
print(is_leap(year))
`````
