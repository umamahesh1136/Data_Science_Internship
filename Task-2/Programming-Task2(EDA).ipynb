## 1.Even Number of Digits

Given an array `nums` of integers, return how many of them contain an **even number of digits**.

### Example 1

**Input**
```text
nums = [12, 345, 2, 6, 7896]
````

**Output**

```text
2
```

**Explanation**

* `12` contains 2 digits (even)
* `345` contains 3 digits (odd)
* `2` contains 1 digit (odd)
* `6` contains 1 digit (odd)
* `7896` contains 4 digits (even)

Only `12` and `7896` have an even number of digits.


### Example 2

**Input**

```text
nums = [555, 901, 482, 1771]
```

**Output**

```text
1
```

**Explanation**

* Only `1771` contains an even number of digits.



### Constraints

* `1 ≤ nums.length ≤ 500`
* `1 ≤ nums[i] ≤ 10⁵`

```
```


`````python
class Solution:
    def findNumbers(self, nums):
        output = 0
        for i in nums:
            if len(str(i)) % 2 == 0:
                output += 1
        return output
`````

`````python
nums = [12,345,2,6,7896]
sol = Solution()
print(sol.findNumbers(nums))
`````

## 2.Number of Good Pairs

Given an array of integers `nums`, return the **number of good pairs**.

A pair `(i, j)` is called **good** if:
- `nums[i] == nums[j]`
- `i < j`

### Example 1

**Input**
```text
nums = [1, 2, 3, 1, 1, 3]
````

**Output**

```text
4
```

**Explanation**

The good pairs are:

* `(0, 3)`
* `(0, 4)`
* `(3, 4)`
* `(2, 5)`


### Example 2

**Input**

```text
nums = [1, 1, 1, 1]
```

**Output**

```text
6
```

**Explanation**

Every possible pair is good because all values are equal.



### Example 3

**Input**

```text
nums = [1, 2, 3]
```

**Output**

```text
0
```

**Explanation**

No two values are equal, so there are no good pairs.

### Constraints

* `1 ≤ nums.length ≤ 100`
* `1 ≤ nums[i] ≤ 100`

```
```


`````python
class Solution:
    def numIdenticalPairs(self, nums: list[int]) -> int:
        count = 0
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                if i < j:
                    if nums[i]==nums[j]:
                        count+=1
        return count
`````

`````python
nums= [1,2,3,1,1,3]
s =  Solution()
s.numIdenticalPairs(nums)
`````

## 3.Defanging an IP Address

Given a valid **IPv4** address, return a **defanged** version of that IP address.

A defanged IP address replaces every period `.` with `[.]`.

### Example 1
**Input:**  
address = `"1.1.1.1"`

**Output:**  
`"1[.]1[.]1[.]1"`

### Example 2
**Input:**  
address = `"255.100.50.0"`

**Output:**  
`"255[.]100[.]50[.]0"`

### Constraints
- The given address is a valid IPv4 address.


`````python
def defangIPaddr(address):
    return address.replace(".", "[.]")
`````

## 4.How Many Numbers Are Smaller Than the Current Number

Given an integer array `nums`, for each `nums[i]`, find how many numbers in the array are **strictly smaller** than it.

That is, for each index `i`, count the number of valid indices `j` such that  
`j != i` and `nums[j] < nums[i]`.

Return the result as an array.

### Example 1
**Input:**  
`nums = [8, 1, 2, 2, 3]`

**Output:**  
`[4, 0, 1, 1, 3]`

**Explanation:**  
- For `8`, there are 4 smaller numbers: `1, 2, 2, 3`  
- For `1`, there are no smaller numbers  
- For `2`, there is 1 smaller number: `1`  
- For `2`, there is 1 smaller number: `1`  
- For `3`, there are 3 smaller numbers: `1, 2, 2`


### Example 2
**Input:**  
`nums = [6, 5, 4, 8]`

**Output:**  
`[2, 1, 0, 3]`



### Example 3
**Input:**  
`nums = [7, 7, 7, 7]`

**Output:**  
`[0, 0, 0, 0]`


### Constraints
- `2 ≤ nums.length ≤ 500`
- `0 ≤ nums[i] ≤ 100`


`````python
class Solution:
   def smallerNumbersThanCurrent(self, nums: list[int]) -> list[int]:
        result = []

        for i in range(len(nums)):
            count = 0
            for j in range(len(nums)):
                if nums[j] < nums[i]:
                    count += 1
            result.append(count)

        return result
`````

`````python
nums = [8, 1, 2, 2, 3]

s = Solution()
result = s.smallerNumbersThanCurrent(nums)

print(result)
`````

## 5.Subtract the Product and Sum of Digits of an Integer

Given an integer number `n`, return the difference between the **product of its digits** and the **sum of its digits**.

### Example 1
**Input:**  
`n = 234`

**Output:**  
`15`

**Explanation:**  
- Product of digits = `2 × 3 × 4 = 24`  
- Sum of digits = `2 + 3 + 4 = 9`  
- Result = `24 − 9 = 15`

### Constraints
- `1 ≤ n ≤ 10⁵`


`````python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        product = 1
        total = 0

        for digit in str(n):
            digit = int(digit)
            product *= digit
            total += digit

        return product - total
`````

`````python
s = Solution()
result = s.subtractProductAndSum(234)
print(result)
`````


## 6.XOR Operation in an Array

You are given two integers `n` and `start`.

Define an array `nums` of length `n` such that:
```

nums[i] = start + 2 * i   (0-indexed)

```

Return the **bitwise XOR** of all elements in the array `nums`.

The `^` symbol represents the bitwise XOR operator.


### Example 1
**Input:**  
`n = 5, start = 0`

**Output:**  
`8`

**Explanation:**  
The array is `[0, 2, 4, 6, 8]`  
`0 ^ 2 ^ 4 ^ 6 ^ 8 = 8`


### Example 2
**Input:**  
`n = 4, start = 3`

**Output:**  
`8`

**Explanation:**  
The array is `[3, 5, 7, 9]`  
`3 ^ 5 ^ 7 ^ 9 = 8`


### Constraints
- `1 ≤ n ≤ 1000`
- `0 ≤ start ≤ 1000`
- `n == nums.length`
```


`````python
class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0

        for i in range(n):
            result ^= start + 2 * i

        return result
`````

`````python
s = Solution()
print(s.xorOperation(5, 0))  # 8
print(s.xorOperation(4, 3))  # 8
`````
