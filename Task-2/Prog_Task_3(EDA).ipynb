
## 1.Maximum Product of Two Elements in an Array

Given an array of integers `nums`, choose **two different indices** `i` and `j` such that `i ≠ j`.

Return the **maximum value** of  
```

(nums[i] - 1) * (nums[j] - 1)

```



### Example 1
**Input:**  
`nums = [3, 4, 5, 2]`

**Output:**  
`12`

**Explanation:**  
Choosing indices `i = 1` and `j = 2` gives:  
```

(4 - 1) * (5 - 1) = 3 * 4 = 12

```


### Example 2
**Input:**  
`nums = [1, 5, 4, 5]`

**Output:**  
`16`

**Explanation:**  
Choosing indices `i = 1` and `j = 3` gives:  
```

(5 - 1) * (5 - 1) = 16

```
### Example 3
**Input:**  
`nums = [3, 7]`

**Output:**  
`12`

### Constraints
- `2 ≤ nums.length ≤ 500`
- `1 ≤ nums[i] ≤ 10³`
```


````python
class Solution:
    def maxProduct(self, nums: list[int]) -> int:
        nums.sort()
        return (nums[-1] - 1) * (nums[-2] - 1)
````

````python
s = Solution()
print(s.maxProduct([3,4,5,2]))  
print(s.maxProduct([1,5,4,5]))  
print(s.maxProduct([3,7]))       
````

## 2.Count Number of Teams

There are `n` soldiers standing in a line. Each soldier is assigned a **unique rating value**.

You need to form a team of **3 soldiers** under the following rules:

- Choose three indices `(i, j, k)` such that `0 ≤ i < j < k < n`
- A team is valid if either:
  - `rating[i] < rating[j] < rating[k]` (strictly increasing), or
  - `rating[i] > rating[j] > rating[k]` (strictly decreasing)

Return the **total number of valid teams** that can be formed.  
A soldier can be part of multiple teams.

### Example 1
**Input:**  
`rating = [2, 5, 3, 4, 1]`

**Output:**  
`3`

**Explanation:**  
Valid teams are:
- `(2, 3, 4)`
- `(5, 4, 1)`
- `(5, 3, 1)`

### Example 2
**Input:**  
`rating = [2, 1, 3]`

**Output:**  
`0`

### Example 3
**Input:**  
`rating = [1, 2, 3, 4]`

**Output:**  
`4`

### Constraints
- `n == rating.length`
- `3 ≤ n ≤ 1000`
- `1 ≤ rating[i] ≤ 10⁵`
- All values in `rating` are unique


````python
class Solution:
    def numTeams(self, rating: list[int]) -> int:
        n = len(rating)
        count = 0

        for i in range(n):
            for j in range(i+1,n): 
                for k in range(j+1,n):
                    if (rating[i] < rating[j] < rating[k]) or \
                       (rating[i] > rating[j] > rating[k]):
                        count += 1
        return count
````

````python
rating = [2, 5, 3, 4, 1]

s = Solution()
result = s.numTeams(rating)

print(result)
````


## 3.Number of Students Doing Homework at a Given Time

You are given two integer arrays `startTime` and `endTime`, and an integer `queryTime`.

The `i`th student started doing their homework at time `startTime[i]` and finished at time `endTime[i]`.

Return the number of students who are doing their homework at time `queryTime`.

More formally, count the number of students such that:
```

startTime[i] ≤ queryTime ≤ endTime[i]

```
(The interval is inclusive.)


### Example 1
**Input:**  
`startTime = [1, 2, 3]`  
`endTime = [3, 2, 7]`  
`queryTime = 4`

**Output:**  
`1`

**Explanation:**  
- Student 1: active from 1 to 3 → not active at 4  
- Student 2: active from 2 to 2 → not active at 4  
- Student 3: active from 3 to 7 → active at 4  


### Example 2
**Input:**  
`startTime = [4]`  
`endTime = [4]`  
`queryTime = 4`

**Output:**  
`1`

**Explanation:**  
The only student is active at time 4.

### Constraints
- `startTime.length == endTime.length`
- `1 ≤ startTime.length ≤ 100`
- `1 ≤ startTime[i] ≤ endTime[i] ≤ 1000`
- `1 ≤ queryTime ≤ 1000`



````python
class Solution:
    def busyStudent(self, startTime: list[int], endTime: list[int], queryTime: int) -> int:
        count  = 0
        for i in range(len(startTime)):
            if startTime[i] <= queryTime <= endTime[i]:
                count += 1
        return count
````

````python
s = Solution()
print(s.busyStudent([1,2,3], [3,2,7], 4))  
print(s.busyStudent([4], [4], 4))          
````


## 4.Number of Steps to Reduce a Number to Zero

Given a non-negative integer `num`, return the number of steps required to reduce it to `0`.

In one step:
- If `num` is **even**, divide it by `2`
- If `num` is **odd**, subtract `1` from it



### Example 1
**Input:**  
`num = 14`

**Output:**  
`6`

**Explanation:**  
```

14 → 7 → 6 → 3 → 2 → 1 → 0

```
### Example 2
**Input:**  
`num = 8`

**Output:**  
`4`

**Explanation:**  
```

8 → 4 → 2 → 1 → 0

```

### Example 3
**Input:**  
`num = 123`

**Output:**  
`12`
### Constraints
- `0 ≤ num ≤ 10⁶`
```


````python
class Solution:
    def numberOfSteps(self, num: int) -> int:
        steps = 0
        while num>0:
            if num%2==0:
                num//=2
            else:
                num-= 1
            steps+=1
        return steps
````

````python
s = Solution()
print(s.numberOfSteps(14))  
print(s.numberOfSteps(8))    
print(s.numberOfSteps(123)) 
````

## 5.Counting Bits

Given an integer `n`, return an array `ans` of length `n + 1` such that for each integer `i` (where `0 ≤ i ≤ n`), `ans[i]` is the number of `1`s in the binary representation of `i`.

### Example 1
**Input:**  
`n = 2`

**Output:**  
`[0, 1, 1]`

**Explanation:**  
```

0 → 0
1 → 1
2 → 10

```

### Example 2
**Input:**  
`n = 5`

**Output:**  
`[0, 1, 1, 2, 1, 2]`

**Explanation:**  
```

0 → 0
1 → 1
2 → 10
3 → 11
4 → 100
5 → 101

```


### Constraints
- `0 ≤ n ≤ 10⁵`

### Follow Up
- Can you solve this in **O(n)** time?
- Can you do it in a **single pass**?
- Can you solve it **without using built-in bit count functions**?
```


````python
class Solution:
    def countBits(self, n: int) -> list[int]:
        ans = []

        for i in range(n+1):
            binary = bin(i)
            ans.append(binary.count('1'))
        return ans
        
````

````python
s = Solution()
print(s.countBits(2))
````
